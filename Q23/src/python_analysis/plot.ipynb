# Visual analysis

I find easier to make data analysis with python, so I provided this notebook which reads from `../data/data.csv` just in case

## LCG generated sequence

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig1 = plt.subplots(figsize=(5, 5))

data = pd.read_csv("../data/data.csv");
n = data.shape[0]
sns.scatterplot(x = data.index, y=data["number"])
```

## Truly random sequence

```python
result = []
arr = [i for i in range(n)]


for i in range(n):
    choosen = np.random.choice(arr)
    arr.remove(choosen)
    result.append(choosen)

fig2 = plt.subplots(figsize=(5, 5))
_ = sns.scatterplot(x = [i for i in range(n)], y=result)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 7))
sns.scatterplot(x = data.index, y=data["number"], ax=ax[0])
ax[0].set_title("LCG")
sns.scatterplot(x = [i for i in range(n)], y=result, ax=ax[1])
ax[1].set_title("Set extraction")
plt.show()
```
