## Python implementation

I first modeled it in python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

G = 9.81 / 0.0254 #gravity
dt = 0.001 
tf = 3
h0 = 10

v = 0
t = 0  
h = h0 

times = []
heights = []

safe = 10
n = 10 
vOld = v
hOld = h
while t < tf: 
    # if not freefalling
    h = h + v*dt - (dt**2)*G/2 
    v = v - G*dt 
    if np.sign(vOld) != np.sign(v) and vOld > 0:
        print(f"{hOld:0.4f}, {h:0.4f}")
    vOld = v
    hOld = h

    if h < 0:
        h = 0 
        n -= 1
        v = np.sqrt(max(2*n*G, 0))
        

    heights.append(h)
    times.append(t)

    t += dt

plt.plot(times, heights)
```
